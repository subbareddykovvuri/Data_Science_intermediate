hence no matter how big the input be, computer only perform 5 operation according to above function. Hence we can say that **number of operation does not depend on the size of input**

````python
# example 2
def foo(n):
  for i in range(n):
    print(i)
````

````python
foo(5)
````

here number of operation that function performs is linearly dependent on the input.

Hence cost of this `foo()` function is **Linear**

````python
# example 3
def foo(n):
  for i in range(n):
    for j in range(n):
      print("WOW!!")
````

````python
foo(3)
````

Here number of operation that function performs is directly proportional to the square of the input.

So cost of this function `foo()` is **quadratic**.


> #### Time complexity is this trend/cost function of number of operations on the input size

For calculating time complexity we only cares about the term with highest power.

```
C(f(x)) = x^n + x ^(n-1) + ..... 1
```
then time complexity will be only `x^n`.
- Constant terms and coefficients are totally ignored

- ## Bubble Sort

<img src="https://d2beiqkhq929f0.cloudfront.net/public_assets/assets/000/015/286/original/Screenshot_2022-09-28_at_10.02.31_AM.png?1664339551">
<img src="https://d2beiqkhq929f0.cloudfront.net/public_assets/assets/000/015/287/original/Screenshot_2022-09-28_at_10.04.49_AM.png?1664339683">

- At each iteration the biggest element of sub-array `[0:n-i]` where `n` is the size of array and `i` is the iteration number, will reach the last position


````python
def bubble_sort(a):
    for i in range(len(a)):
        for j in range(i, len(a)-1):
            if a[j] > a[j+1]:
                a[j], a[j+1] = a[j+1], a[j]

    return a
````

````python
a = [5,1,2,4,7,3]
bubble_sort(a)
````
