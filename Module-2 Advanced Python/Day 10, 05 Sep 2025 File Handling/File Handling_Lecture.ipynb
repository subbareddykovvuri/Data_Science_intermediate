```python
import threading
import time

#Line 1: Python threading demo
#Line 2: Threads share resources
#Line 3: Locks prevent race conditions
#Line 4: This is an example file
#Line 5: Each thread will read this
#Line 6: Notice the difference with lock
#Line 7: Without lock output gets jumbled
#Line 8: With lock it is synchronized

def read_file(fname,thread_name):
  with open(fname,'r') as f:
    for line in f:
      print(f"{thread_name} --> {line.strip()}")
      time.sleep(0.01)

fname = "sample.txt"

t1 = threading.Thread(target=read_file,args = (fname,"Thread-1"))
t2 = threading.Thread(target=read_file,args = (fname,"Thread-2"))
t3 = threading.Thread(target=read_file,args= (fname,"Thread-3"))

t1.start()
t2.start()
t3.start()

t1.join()
t2.join()
t3.join()
```

```python
#with Lock

import threading

#Line 1: Python threading demo
#Line 2: Threads share resources
#Line 3: Locks prevent race conditions
#Line 4: This is an example file
#Line 5: Each thread will read this
#Line 6: Notice the difference with lock
#Line 7: Without lock output gets jumbled
#Line 8: With lock it is synchronized

lock = threading.Lock()

def read_file(fname,thread_name):
  with open(fname,'r') as f:
    for line in f:
      with lock:
        print(f"{thread_name} --> {line.strip()}")

fname = "sample.txt"

t1 = threading.Thread(target=read_file,args = (fname,"Thread-1"))
t2 = threading.Thread(target=read_file,args = (fname,"Thread-2"))
t3 = threading.Thread(target=read_file,args= (fname,"Thread-3"))

t1.start()
t2.start()
t3.start()

t1.join()
t2.join()
t3.join()
```

```python
file = open("sample.txt", "r+")

buffer = file.readline()

while buffer: # n lines = 1 block of memory
    print(buffer, end = "")
    buffer = file.readline()

file.close()
```
