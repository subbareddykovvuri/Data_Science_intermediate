**Q2. Solving a Business Problem**

Choose the correct order for the following statements to be either True or False.

S1: You need to understand the business problem well before starting to build a solution.

S2: You don’t need to have a clear idea of what you’re trying to achieve. Just a rough mind map is enough.

S3: It’s all about delivering satisfactory results to the clientele at the end.

----

True; True; False

False; True; False

True; False; True

True; False; False

The correct order for these statements is **True, False, True**.

Here is a quick breakdown of why:

* **S1 (True):** Jumping into a solution without understanding the business problem is like trying to prescribe medicine without a diagnosis. You risk wasting time, money, and resources on a product that doesn’t actually fix the pain point.
* **S2 (False):** While a mind map is a great starting point for brainstorming, it isn't enough for a final project scope. You need **Clear, Measurable Objectives**. Without a specific "North Star," projects tend to suffer from scope creep and lack of direction.
* **S3 (True):** In any professional setting, the ultimate measure of success is whether the solution provides value and meets the expectations of the stakeholders or clientele.

---

### The Problem-Solving Workflow

In a professional environment, the process usually looks like this:

| Phase | Goal |
| --- | --- |
| **Discovery** | Deep dive into the business "why" (Validates S1). |
| **Definition** | Turning rough ideas into concrete requirements (Refutes S2). |
| **Delivery** | Ensuring the final output solves the client's needs (Validates S3). |

**Q3. Metric Change**

Which of the following questions is not related to metric change in product diagnosis?

----


Why has the conversion ratio of free users to subscribers decreased by 10% over the past years?

Why are the customers not visiting the food courts these days?

Why is the estimated time for cab arrival 3 mins higher than normal?

Why is the average sales of grocery items down by 4.5% this month?


The correct answer is: **Why are the customers not visiting the food courts these days?**

### **Why this is the correct choice:**

In product management and data analytics, a **Metric Change** question typically includes three specific components:

1. **A specific metric:** (e.g., conversion ratio, arrival time, sales).
2. **A direction of change:** (e.g., decreased, higher, down).
3. **A magnitude or time frame:** (e.g., 10%, 3 mins, this month).

The food court question is a **broad discovery question**. It lacks a specific data point or a measurable baseline (e.g., "Why is footfall down by 20%?"). It is a general inquiry into behavior rather than a diagnosis of a shifting data trend.

---

### **Evaluating the other options:**

* **Conversion ratio decreased by 10%:** This is a classic **Retention/Monetization** metric change.
* **Cab arrival 3 mins higher:** This is an **Operational Efficiency** metric change.
* **Average sales down by 4.5%:** This is a **Revenue/Growth** metric change.

### **The Metric Diagnosis Framework**

When a metric change is identified, analysts usually follow a "Root Cause Analysis" (RCA) tree to find the source.

| Question Type | Focus | Example |
| --- | --- | --- |
| **Metric Change** | Quantitative "What happened?" | "Why is DAU down 5%?" |
| **Problem Discovery** | Qualitative "Why is this happening?" | "Why don't people like the UI?" |

This question falls under **Product Strategy and Prioritization**. It is asked during the discovery or planning phase to decide *what* to build.

* **"Which metric would you observe...?"** – This is a direct request for a **North Star Metric** or primary success indicator.

To visualize where these questions fit, consider the standard product development flow:

1. **Strategy Phase:** "Should we improve or build new?" (The excluded question).
2. **Execution Phase:** "How do we ensure accuracy?"

**Q5. Internal Factor**

Which of these is not a potential internal factor that can cause a metric change?

---

Geography

Competitors

User Demographics

Platform

The correct answer is: **Competitors**.

### **Why this is the correct choice:**

When diagnosing why a metric (like sales or user engagement) has changed, we categorize factors into **Internal** (things the company controls or owns) and **External** (things happening in the outside world).

* **Competitors** are an **External Factor**. You cannot control a competitor’s new feature launch, their aggressive pricing strategy, or their marketing spend, even though these things directly impact your metrics.

---

### **Breakdown of Internal Factors:**

The other options are considered internal because they relate to the **segmentation of your own data** and product environment:

* **Geography:** While the physical location is "outside," in business analytics, this refers to your **internal market segments**. If your app crashes specifically in India but works in the US, that is an internal technical or operational issue.
* **User Demographics:** This refers to the specific **sub-groups of your existing user base** (e.g., Age, Gender, Occupation). A shift here usually points to an internal change in who your product is attracting or retaining.
* **Platform:** This refers to the **technical medium** (iOS, Android, Web). If a metric drops only on Android, it’s almost always an internal bug or a versioning issue within your control.

### **Summary Table**

| Category | Factors | Control Level |
| --- | --- | --- |
| **Internal** | Tech bugs, UI changes, Geography/Demographics, Marketing campaigns | **High** |
| **External** | **Competitors**, Government regulations, Seasonality (holidays), Global Pandemics | **Low/None** |

**Q6. Importance of Defining Success**

Defining success for a new product or feature is important to..?

---

Validate the product/feature

Generate more revenue

Attract new customers

Estimate the cost

The correct answer is: **Validate the product/feature**.

### **Why this is the primary goal:**

While generating revenue and attracting customers are often the *desired outcomes* of a business, the specific act of **defining success** is a measurement exercise. You define success to determine if the product or feature actually solves the problem it was intended to fix.

* **Validation:** By setting clear success metrics (like user retention or click-through rates), you can objectively prove whether the feature "works" or if the hypothesis behind it was correct. Without this, you might continue investing in a product that users don't actually find valuable.

---

### **Breakdown of the other options:**

* **Generate more revenue / Attract new customers:** These are specific *types* of business goals, but they aren't the **reason** we define success. You can define a project as "successful" even if it doesn't make money (e.g., a security update or a "dark mode" feature) as long as it meets its specific validation criteria.
* **Estimate the cost:** This is a budgeting and resource planning task that happens *before* or during the proposal phase, rather than the process of evaluating the success of the outcome.

### **The Validation Loop**

Defining success creates a feedback loop:

1. **Hypothesis:** "Adding a 'Share' button will increase virality."
2. **Metric:** Define success as a **5% increase in referrals**.
3. **Analysis:** If referrals increase by 5%, the feature is **Validated**. If not, you pivot.
