
**Q1. Test the Training Program**

You are appointed as a Data Analyst for a training program deployed by the Government of India.

The participantsâ€™ skills were tested before and after the training using some metrics on a scale of 10.

before = [2.45, 0.69, 1.80, 2.80, 0.07, 1.67, 2.93, 0.47, 1.45, 1.34]   

after = [7.71, 2.17, 5.65, 8.79, 0.23, 5.23, 9.19, 1.49, 4.56, 4.20]
Conduct an appropriate test to assess a statistically significant increase in the average skill score after the training program, and then answer the below questions accordingly.

**Note:** Perform the test at alpha = 5%.

---

**Choose the correct answer from below, please note that this question may have multiple correct answers**

T statistic = 5.11

There is no effect of the training.

There is a positive effect of the training.

We fail to reject the null hypothesis.

We reject the null hypothesis.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# 1. Data Setup
before = np.array([2.45, 0.69, 1.80, 2.80, 0.07, 1.67, 2.93, 0.47, 1.45, 1.34])
after = np.array([7.71, 2.17, 5.65, 8.79, 0.23, 5.23, 9.19, 1.49, 4.56, 4.20])

# 2. Normality Check (Shapiro-Wilk Test)
# A Paired T-test assumes the differences between pairs are normally distributed
differences = after - before
shapiro_stat, shapiro_p = stats.shapiro(differences)

# 3. Paired T-test
# We use a 'greater' alternative because we are testing for an INCREASE
t_stat, p_val = stats.ttest_rel(after, before, alternative='greater')

# 4. Results Output
print("--- Statistical Test Results ---")
print(f"Shapiro-Wilk Normality p-value: {shapiro_p:.4f}")
print(f"T-statistic: {t_stat:.4f}")
print(f"P-value (One-tailed): {p_val:.4f}")

alpha = 0.05
if p_val < alpha:
    print("\nResult: Reject the Null Hypothesis.")
    print("Conclusion: There is a statistically significant increase in skill scores.")
else:
    print("\nResult: Fail to reject the Null Hypothesis.")
    print("Conclusion: There is no significant increase in skill scores.")

# 5. Visualization (Optional but recommended for analysts)
plt.figure(figsize=(8, 5))
plt.boxplot([before, after], tick_labels=['Before Training', 'After Training'])
plt.title('Comparison of Skill Scores Before and After Training')
plt.ylabel('Score (Scale of 10)')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import numpy as np
from scipy import stats

# Input data
female_scores = [25, 30, 45, 49, 47, 35, 32, 42]
male_scores = [45, 47, 25, 22, 29, 32, 27, 28, 40, 49, 50, 33]

# Perform independent t-test (two-tailed)
t_stat, p_two_tailed = stats.ttest_ind(female_scores, male_scores)

# Calculate one-tailed p-value for (female > male)
# Since mean_female (38.125) > mean_male (35.583), t_stat is positive.
# One-tailed p-value = Two-tailed p-value / 2
p_value = p_two_tailed / 2

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value (one-tailed): {p_value:.3f}")

# Comparison with alpha
alpha = 0.02
if p_value <= alpha:
    print("Significant evidence: Reject H0")
else:
    print("No significant evidence: Fail to reject H0")
```

**Q6. Marital Status and Drinking**

A national survey was conducted to obtain information on the alcohol consumption patterns of U.S. adults by marital status.
A random sample of 1772 residents, aged 18 and older, yielded the data displayed in Table below:
Screenshot_2023-04-27_at_1.36.46_PM.png


Test whether Marital status and alcohol consumption are associated with a 5% significance level.


---


Choose the correct option below :
p_value = 4.9513e^-15, Marital status and alcohol consumption are associated.

p_value = 4.9513e^-15, Marital status and alcohol consumption are not associated.

p_value = 3.9251e^-18, Marital status and alcohol consumption are not associated.

p_value = 3.9251e^-18, Marital status and alcohol consumption are associated.

```python
import numpy as np
from scipy.stats import chi2_contingency

# Replace these values with the actual data from your table
# Each row represents a Marital Status, each column an Alcohol Consumption level
data = [
    [67, 213, 74],
    [411, 633, 129],
    [85, 51, 7],
    [27, 60, 15]
]

# Convert to numpy array
obs = np.array(data)

# Perform Chi-square test
chi2, p_value, dof, expected = chi2_contingency(obs)

print(f"Chi-square Statistic: {chi2:.4f}")
print(f"P-value: {p_value}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies Table:")
print(expected)

# Interpret results at 5% significance level
alpha = 0.05
if p_value < alpha:
    print("\nResult: Reject the Null Hypothesis (H0).")
    print("Conclusion: Marital status and alcohol consumption are associated.")
else:
    print("\nResult: Fail to reject the Null Hypothesis (H0).")
    print("Conclusion: No significant evidence of an association.")
```

**Q8. Time spent on website**

Suppose you are interested in the distribution of time spent on a website, by it's users. You expect that:

20% of users spend less than 5 minutes,
50% spend between 5 and 10 minutes, and
30% spend more than 10 minutes.
After collecting data from 200 users, you find that

30 users spent less than 5 minutes,
85 users spent between 5 and 10 minutes, and
85 users spent more than 10 minutes.
Conduct an appropriate test to see if the distribution of browsing times matches your expectations at a 5% significance level. Choose the correct option from below:



a)

P-value: 0.0005

We fail to reject the null hypothesis

Thus concluding that the distribution of browsing times matches expectations.

b)

P-value: 0.0005

The null hypothesis is rejected,

Thus concluding that the distribution of browsing times does not match expectations.

c)

P-value: 0.33

We fail to reject the null hypothesis

Thus concluding that the distribution of browsing times matches expectations.

d)

P-value: 0.33

The null hypothesis is rejected,

Thus concluding that the distribution of browsing times does not match expectations.

```python
import numpy as np
from scipy import stats

# Given data
n = 200
observed = np.array([30, 85, 85])
expected_proportions = np.array([0.20, 0.50, 0.30])

# Calculate expected frequencies
expected = n * expected_proportions

# Perform Chi-square Goodness of Fit test
chi2_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)

# Degrees of freedom
df = len(observed) - 1

# Significance level
alpha = 0.05

print(f"Observed Frequencies: {observed}")
print(f"Expected Frequencies: {expected}")
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"Degrees of Freedom: {df}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    result = "Reject the null hypothesis (H0)."
    conclusion = "The distribution of browsing times does not match the expectations."
else:
    result = "Fail to reject the null hypothesis (H0)."
    conclusion = "The distribution of browsing times matches the expectations."

print(f"Result: {result}")
print(f"Conclusion: {conclusion}")
```

**Q10. Distribution of smartphone brands**

A Mobile Retail store owner is interested in the distribution of popular smartphone brands among a group of 200 people.

They expect that 30% of people would prefer Brand A, 40% would prefer Brand B and 30% would prefer Brand C.

However, upon surveying the group, the results are as follows: 70 prefer Brand A, 80 prefer Brand B, and 50 prefer Brand C.

Conduct an appropriate test to see if the distribution of preferences matches the store owner's expectations at a 5% significance level.

Choose the correct option below:

a)

P-value: 0.2048

We fail to reject the null hypothesis

Thus concluding that the observed distribution matches the expectations.

b)

P-value: 0.2048

The null hypothesis is rejected

Thus concluding that the observed distribution does NOT match the expectations.

c)

P-value: 0.1888

We fail to reject the null hypothesis

Thus concluding that the observed distribution matches the expectations.

d)

P-value: 0.1888

The null hypothesis is rejected

Thus concluding that the observed distribution does not match the expectations

---



```python
import numpy as np
from scipy import stats

# Given data
n = 200
observed = np.array([60, 80, 60])

expected = np.array([70, 80, 50])

# Perform Chi-square Goodness of Fit test
chi2_stat, p_value = stats.chisquare(f_obs=observed, f_exp=expected)

# Degrees of freedom
df = len(observed) - 1

# Significance level
alpha = 0.05

print(f"Observed Frequencies: {observed}")
print(f"Expected Frequencies: {expected}")
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"Degrees of Freedom: {df}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    result = "Reject the null hypothesis (H0)."
    conclusion = "the distribution of preferences does not matches the store owner's expectations"
else:
    result = "Fail to reject the null hypothesis (H0)."
    conclusion = "the distribution of preferences matches the store owner's expectations"

print(f"Result: {result}")
print(f"Conclusion: {conclusion}")
```
